```python
customers = pd.read_csv('/content/drive/MyDrive/Dataset/Customers.csv')
products = pd.read_csv('/content/drive/MyDrive/Dataset/Products.csv')
transactions = pd.read_csv('/content/drive/MyDrive/Dataset/Transactions.csv')
```

```python
print(customers.columns)
```

```python
print(products.columns)
```

```python
print(transactions.columns)
```

```python
data = transactions.merge(customers, on='CustomerID').merge(products, on='ProductID')
```

```python
top_products = data.groupby('ProductName')['TotalValue'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar', color='skyblue', title='Top 10 Selling Products')
plt.ylabel('Total Revenue')
plt.xlabel('Product Name')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
low_products = data.groupby('ProductName')['TotalValue'].sum().sort_values(ascending=True).head(10)
plt.figure(figsize=(10, 6))
low_products.plot(kind='bar', color='lightcoral')
plt.title('Bottom 10 Selling Products by Revenue')
plt.ylabel('Total Revenue')
plt.xlabel('Product Name')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
